# Mean

```python
#Libraries
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("Baseball.csv")
df.head()
```

Mean


```python
#Mean for Runs Scored(RS)
df.RS.mean()
```

```python
#Mean for Runs Scored(RS) by Arizona Team(ARI)
df.loc[df.Team == "ARI"].RS.mean()
```

```python
#Mean for Runs Scored(RS) by Arizona Team(ARI) since 2005
df.loc[(df.Team == "ARI") & (df.Year >= 2005)]
```

```python
#What is the mean of Runs Allowed(RA) by Chicago Team(CHC) until 2007
df.loc[(df.Team == "CHC") & (df.Year <2007)]
```

```python
df.loc[(df.Team == "CHC") & (df.Year <2007)].RA
```

```python
df.loc[(df.Team == "CHC") & (df.Year <2007)].RA.mean()
```

# Median

```python
#Mean and Median of Wins(W) by all the Teams
print(df.W.mean())
print(df.W.median())
```

```python
#Median of the Wins(W) by the Baltimore Team(BAL) until 2000
df.loc[(df.Team == "BAL") & (df.Year <= 2000)]
```

```python
df.loc[(df.Team == "BAL") & (df.Year <= 2000)].W.median()
```

# Mode

```python
#Mode, Mean and Median of OBP
print(df.OBP.mean())
print(df.OBP.median())
print(df.OBP.mode())
```

```python
#Mode of OBP during the Year 2010
df.loc[df.Year == 2010]
```

```python
df.loc[df.Year == 2010].OBP.mode()
```

# Correlation

```python
#Pick variables for Correlation
df_correlation = df[["RS","RA","W"]]
df_correlation
```

```python
#Creating the Correlation Matrix
df_correlation.corr()
```

```python
#Using seaborn create a heatmap
sns.heatmap(data=df_correlation.corr(),
            annot = True,
            fmt = ".2g",
            center = 0,
            cmap = "coolwarm"
)
```

```python
#Create a Correlation Matrix between OBP, SLG and BA. Also create a heatmap for the same.
df_correlation_new = df[["OBP","SLG","BA"]]
df_correlation_new.head(1)
```

```python
df_correlation_new.corr()
```

```python
sns.heatmap(data = df_correlation_new.corr(),
            annot = True,
            center = .7,
            fmt = ".1g",
            cmap = "crest")
```

# Standard Deviation

```python
#In order to calculate Standard Deviation, Mean is required for the given set of values
print(df.OOBP.mean())
```

```python
print(df.OOBP.std())
```

```python
# Calculating the Standard Deviation of BA
print(df.BA.mean())
print(df.BA.std())
```
